```python
# The data is from LendingClub.com with information about lending to individuals.
# as an investor we will try to predict that if we lend money then whats the probability of person paying his loan back
```

```python
# importing libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.info()
```

```python
# the data
df.head()
```

```python
#  histogram of two FICO distributions on top of each other, one for each credit.policy outcome

plt.figure(figsize=(12,6))
plt.tight_layout()
sns.set_style('darkgrid')
df[df['credit.policy'] == 1]['fico'].plot.hist(bins=30,label='Credit policy =1')
df[df['credit.policy'] == 0]['fico'].plot.hist(bins=30,label='Credit policy =0')
plt.xlabel('FICO')
plt.legend()
```

```python
# similar plot by not fully paid.

plt.tight_layout()
sns.set_style('darkgrid')
df[df['not.fully.paid']==0]['fico'].plot.hist(bins=30,label = 'Not Fully Paid')
df[df['not.fully.paid']==1]['fico'].plot.hist(bins=30,label = 'Fully Paid')
plt.legend()
```

```python
# Count plot according to purpose of loan and full payment status

sns.countplot(data = df, x='purpose',hue='not.fully.paid')
```

```python
#  the trend between FICO score and interest rate
plt.figure(figsize=(12,7))
plt.tight_layout()
sns.jointplot(data = df, x = 'fico', y = 'int.rate', color='purple')
```

```python
#  lmplots to see if the trend differed between not.fully.paid and credit.policy

sns.lmplot(data = df , x = 'fico', y = 'int.rate',col='not.fully.paid',hue='credit.policy')
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(df,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

```python
final_data.head()
```

```python
# training and testing the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
# building up decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
# Random forest model 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_predict = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_predict))
```

```python
print(confusion_matrix(y_test,rfc_predict))
```
